* The methods used to change page should be chainable, so you can call them one
after the other like this: p.nextPage().nextPage()

### 2) Coffee Shop

Write a class called CoffeeShop, which has **three instance variables**:
1. name: a string (basically, of the shop)
2. menu: a list of items (of dict type), with each item containing the item (name of
the item), type (whether a food or a drink) and price.
3. orders: an empty list

and **seven methods**:

1. **add_order**: adds the name of the item to the end of the orders list if it exists on
the menu, otherwise, return "This item is currently unavailable!"
2. **fulfill_order**: if the orders list is not empty, return "The {item} is ready!". If
the orders list is empty, return "All orders have been fulfilled!"
3. **list_orders**: returns the item names of the orders taken, otherwise,
an empty list.
4. **due_amount**: returns the total amount due for the orders taken.
5. **cheapest_item**: returns the name of the cheapest item on the menu.
6. **drinks_only**: returns only the item names of type drink from the menu.
7. **food_only**: returns only the item names of type food from the menu.

**IMPORTANT**: Orders are fulfilled in a **FIFO** (first-in, first-out) order.

```python
class coffeeShop:
    def __init__(self,name,menu):
        self.name = name
        self.menu = menu
        self.orders = []
        self.price = 0.0

        
    def add_order(self, order):
        '''Add order if the item exist in the menu'''
        
        for i in self.menu:
            if order == i["item"]:
                self.orders.append(order)
                print("Order added!")
                
        if order not in self.orders:
            print('This item is currently unavailable')
            
    
    def fulfill_order(self):
        '''pop an order if orders is not empty [First in First out]'''
        
        if self.orders:
            popped_order = self.orders.pop(0)
            print(f"The {popped_order} is ready!")
        else:
            print("All orders have been fulfilled!")
            

    def list_order(self):
        '''return the orders in the order list'''
        
        print(self.orders)
        
    def due_amount(self):
        '''return the total price of the items in the order list'''
        
        for i in self.orders:
            for j in self.menu:
                if i == j['item']:
                    self.price += j['price']
                    
        return self.price
    
    
    def cheapest_item(self):
        '''return the cheapest item in the menu'''
        
        cheapest_price = self.menu[0]['price']
        cheapest_item = self.menu[0]['item']
        
        for i in self.menu:
            if i['price'] < cheapest_price :
                cheapest_item = i['item']
                cheapest_price = i['price']
                
        return cheapest_item
    
    
    def drinks_only(self):
        '''return the drinks only in the menu'''
        
        drinks = []
        for i in self.menu:
            if i["type"] == "drink":
                drinks.append(i["item"])
                
        return drinks
    
    
    def food_only(self):
        '''return the food only in the menu'''
        
        food = []
        for i in self.menu:
            if i["type"] == "food":
                food.append(i["item"])
        
        return food

        
        
menu = [
    {"item": "cinnamon roll", "type": "food", "price": 25},
    {"item": "iced coffee", "type": "drink", "price": 20},
    {"item": "tuna sandwitch", "type": "food", "price": 27},
    {"item": "lemon", "type": "drink", "price": 21},
    {"item": "ham and cheese sandwich", "type": "food", "price": 30},
    {"item": "orange juice", "type": "drink", "price": 18},
    {"item": "bacon and egg", "type": "food", "price": 28},
    {"item": "pineapple", "type": "drink", "price": 23},
]

c = coffeeShop('retro',menu)
c.add_order('iced coffee')
c.add_order('ham and cheese sandwich')
c.add_order('cinnamon roll')
c.add_order('bacon and egg')
c.add_order('pineapple')
c.add_order('cake')

c.fulfill_order()
c.fulfill_order()

c.list_order()
c.due_amount()
#c.cheapest_item()
#c.drinks_only()
#c.food_only()
```
